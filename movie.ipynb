```python
import pandas as pd
import numpy as np
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
md=pd.read_csv('movies.csv')
md.head()
md.shape
features=['genres','keywords','tagline','cast','director']
print(features)
for f in features:
  md[f]=md[f].fillna('')
cf =md['genres']+' '+ md['keywords']+' '+ md['tagline']+' '+ md['cast']+' '+md['director']
vector = TfidfVectorizer()
fv = vector.fit_transform(cf)
print(fv)
```

```python
s = cosine_similarity(fv)
print(s)
print(s.shape)
m=input('enter the movie name')
```

```python
l = md['title'].tolist()
suggested = difflib.get_close_matches(m,l)
print(suggested)
```
